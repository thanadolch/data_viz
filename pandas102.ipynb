ต่อแกน X



*   จับ 2 ตารางมาต่อกันเลย (merge)
*   เลือกมาเพิ่มเฉพาะ column (map)





````
data_province = data_covid[['No.','announce_date','province_of_onset']]
````

````
data_human = data_covid[['No.','age','sex','nationality']]
````

````
full_table1 = data_human.merge(data_province)
full_table1.head()
````

````
human2 = data_human.sort_values('age')
````

````
full_table2 = human2.merge(data_province)
full_table2
````

````
data_human2_renamed = human2.rename(columns={'No.':'patientNumber'})
data_human2_renamed
````

````
full_table3 = data_human2_renamed.merge(data_province,left_on='patientNumber',right_on='No.')
full_table3
````

## map() เลือกมาเฉพาะบาง column มาแปะเพิ่มเข้าไป

````
import statsmodels.api as sm
from statsmodels.formula.api import ols
````

````
covid = ols('age ~ C(province_of_onset)',data=data_covid).fit()
table2 = sm.stats.anova_lm(covid, typ=2)
print(table2)
````

````
import statsmodels.api as sm
from statsmodels.formula.api import ols
moore = sm.datasets.get_rdataset("Moore", "carData",cache=True) # load data
data = moore.data

data = data.rename(columns={"partner.status":"partner_status"}) # make name pythonic 

moore_lm = ols('conformity ~ C(fcategory, Sum)*C(partner_status, Sum)',data=data).fit() 

table = sm.stats.anova_lm(moore_lm, typ=2) # Type 2 ANOVA DataFrame

print(table)
````
