#การแสดงผลที่ดูการกระจายของข้อมูล
### Scatter
### Scatter 3D

```
import pandas as pd
from matplotlib import pyplot as plt
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
```

scatter-defualt

```
plt.scatter(df['PetalLength'],df['PetalWidth'])
```

```
set(df['Name'])
```

```
df2 = df.replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'])
```

```
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'],alpha=0.5)
```

##ใช้ขนาดของ marker แสดงค่าของ feature

```
plt.scatter(df['PetalLength'],df['PetalWidth'],s=20*df['SepalWidth'],c=df2['Name'],alpha=0.5)
```

##ตัวแปร Marker ใช้ในการเปลี่ยนหน้าตาของ marker

```
plt.scatter(df['PetalLength'],df['PetalWidth'],s=20*df['SepalWidth'],c=df2['Name'],alpha=0.5,marker='x')
```

##การ plot กราฟหลายๆอันซ้อนกัน

```
plt.scatter(df['PetalLength'][:50],df['PetalWidth'][:50],s=20*df['SepalWidth'],c='y',alpha=0.5,marker='$a$')
plt.scatter(df['PetalLength'][50:100],df['PetalWidth'][50:100],s=20*df['SepalWidth'],c='c',alpha=0.5,marker='$b$')
plt.scatter(df['PetalLength'][100:],df['PetalWidth'][100:],s=20*df['SepalWidth'],c='r',alpha=0.5,marker='$c$')
```
