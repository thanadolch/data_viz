หาจำนวนใน column 'Unnamed: 8' ที่มีค่า null แล้วลบทิ้งเพราะเป็น row ที่เกินมา

```
test_in_class[test_in_class['Unnamed: 8'].isnull()]
test_in_class2 = test_in_class.drop(index=range(89,100))
test_in_class2
```

1. เลือกcolumnที่ต้องการใช้
2. ลบแถวที่ไม่จำเป็นคือแถวที่ 0 
3. เลือกค่าในตารางที่ไม่ใช้ QC
4. เปลี่ยนtypeของฟังชั่ทั้งหมดในแถว FORM8 (เข้าพบ ทปษ.) ให้เป็น float

```
test_in_class_real = test_in_class2[['เลขที่','รหัสประจำตัว','ชื่อ','สาขา','Unnamed: 8','FORM8 (เข้าพบ ทปษ.)']]
test_in_class_real = test_in_class_real.drop(index=0)
test_in_class_real = test_in_class_real[test_in_class_real['FORM8 (เข้าพบ ทปษ.)']!='QC']
test_in_class_real1 = test_in_class_real['FORM8 (เข้าพบ ทปษ.)'].astype(float)
test_in_class_real2 = test_in_class_real['Unnamed: 8'].astype(float)
test_in_class_real2
```

นำคอลัมน์ FORM8 (เข้าพบ ทปษ.) Unnamed: 8 มาบวกกันแล้วตั้งค่าตัวแปรรับเป็น pip

```
pip = test_in_class_real1 + test_in_class_real2
pip
```

นำค่าในตัวแปร pip มาเติมในตาราง test_in_class_reeal แล้วตั้งชื่อ column ว่า total

```
test_in_class_real['total'] = pip
test_in_class_real
```
