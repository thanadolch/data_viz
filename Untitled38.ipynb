```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

```
pd.plotting.parallel_coordinates(df,'Name', color=('#5562','#615586','#F50909')) #สีแบบรหัสในHTML
```

```
pd.plotting.parallel_coordinates(df,'Name', color=('r','g','b')) #สีแบบตัวย่อ
```

```
pd.plotting.parallel_coordinates(df,'Name', color=((1,0,0),(0,1,0),(0,0,1))) #RGB value ซึ่ง RGBต้องมีค่าตั้งแต่ 0,1
```

```
pd.plotting.parallel_coordinates(df,'Name', colormap='cividis')
```

Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็นnumpy array

```
np_data = df.iloc[:,:-1].to_numpy()
```

visualize array ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure

plt.imshow(np_data[:10],interpolation='nearest')
```

การใช้ subplot

```
plt.subplot(1,2,1) #(มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไร่)
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma')
plt.subplot(1,2,2) #(มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไร่)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')
```

```
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma',alpha = 0.7)
```
